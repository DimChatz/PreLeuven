### Conservation function

```python
# Calculation of frequencies in a dataset or a specific column
def freqAll(dataset,n):
    freq = np.zeros((4, 1))
    if n==2:
        for i in range(dataset.shape[0]):
            for j in range(dataset.shape[1]):
                if dataset[i][j]=="A":
                    freq[0]+=1
                if dataset[i][j]=="T":
                    freq[1]+=1
                if dataset[i][j]=="C":
                    freq[2]+=1
                if dataset[i][j]=="G":
                    freq[3]+=1
        for i in range(dataset.shape[0]):
            freq[i] = freq[i]/(dataset.shape[0]*dataset.shape[1])
    if n==1:
         for i in range(dataset.shape[0]):
            if dataset[i][j]=="A":
                freq[0]+=1
            if dataset[i][j]=="T":
                freq[1]+=1
            if dataset[i][j]=="C":
                freq[2]+=1
            if dataset[i][j]=="G":
                freq[3]+=1
        for i in range(dataset.shape[0]):
            freq[i] = freq[i]/dataset.shape[0]
    return freq

# Calculation relative entropy
def RE(column, dataset,n1,n2):
    freq1 = freqAll(column,n1)
    freq2 = freqAll(dataset,n2)
    relativ = 0
    for i in range(dataset.shape[0]):
        relativ = relativ + freq1[i]*log(freq1[i]/(0.5*freq1[i]+0.5*freq2[i]))
    return relativ


# Conservation calculation 2 - Jensen-Shannon Divergence
def conservJSD(column):
    return JSD = 0.5*RE(column,dataset,1,2) + 0.5*RE(dataset,column,2,1)
```
